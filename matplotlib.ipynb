```python

#Basic line plot
months = np.array(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
sales = np.array([200,220,250,275,300,320,310,295,280,260,250,240])  # Added one more value

plt.figure(figsize=(6,4))
plt.plot(months, sales, marker='o', linestyle='-', color='b')
plt.title('Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
#Histogram
np.random.seed(0) # For Reproductibility
data=np.random.normal(loc=0,scale=1,size=1000) #Generate random data with normal distribution
plt.figure(figsize=(6,4))
plt.hist(data,bins=30,color='purple',edgecolor='black')
plt.title('Random Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Scatter plot
np.random.seed(1)
x=np.random.rand(50)*100 #Random x values
y=np.random.rand(50)*100 #Random y values

plt.figure(figsize=(6,4))
plt.scatter(x,y,color='red',alpha=0.7)
plt.title('Random Scatter Plot')
plt.xlabel(' X Values')
plt.ylabel('y Values')
plt.grid(True)
plt.show()
```

```python
#Bar plot
categories=np.array(['A','B','C','D','E'])
values=np.array([10,20,15,25,30])

plt.figure(figsize=(6,4))
plt.bar(categories,values,color=['blue','green','red','purple','orange'])
plt.title('Category value')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(False)
plt.show()
```

```python
#Box plot
np.random.seed(2)
group1=np.random.normal(loc=0,scale=1,size=100)
group2=np.random.normal(loc=1,scale=1.5,size=100)

plt.figure(figsize=(6,4))
plt.boxplot([group1,group2],labels=['Group1','Group2'])
plt.title('Box Plot of Two Groups')
plt.xlabel('Group')
plt.ylabel('Value')
plt.grid(True)
plt.show()
                        
                        
```
