```python
!pip install matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(2)
group1=np.random.normal(loc=0,scale=1,size=100)
group2=np.random.normal(loc=1,scale=1.5,size=100)

plt.figure(figsize=(6,4))
plt.boxplot([group1,group2],labels=['Group1','Group2'])
plt.title('Box Plot of Two Groups')
plt.xlabel('Group')
plt.ylabel('Value')
plt.grid(True)
plt.show()

```

```python
#Heatmap (Correlation Matrix)
np.random.seed(3)
data_matrix=np.random.rand(10,10)
plt.figure(figsize=(6,4))
plt.imshow(data_matrix,cmap='coolwarm',interpolation='none')
plt.title('Heatmap of Random Data Matrix')
plt.colorbar(label='Value')
plt.show()
```

```python
# Creating Subplot
x = np.linspace(0.10, 100, 100)  # 100 values from 0.10 to 100
y1 = np.sin(x)
y2 = np.cos(x)

# CREATE A FIGURE
plt.figure(figsize=(6, 2))

# First Subplot: Sine wave (assigned to the first plot of the grid)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
plt.plot(x, y1, 'b-')  # Line plot for sine wave
plt.title=('Sine Wave')

# Second Subplot: Cosine wave (assigned to the second plot of the grid)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
plt.plot(x, y2, 'r--')  # Line plot for cosine wave
plt.title=('Cosine Wave')

# Adjust layout to avoid overlap
plt.tight_layout()

# Display the plots
plt.show()
```
